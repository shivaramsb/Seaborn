```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
exercise = sns.load_dataset("exercise")
```

**1. Display the first 5 rows of the dataset.**

```python
print(exercise.head())
```

 **2. Create a bar plot showing the average pulse rate by diet.**

```python
sns.barplot(x="diet", y="pulse", data=exercise)
plt.title("Average Pulse Rate by Diet")
plt.show()
```

**3. Create a box plot to show the distribution of pulse rates across different kinds of exercises.**

```python
sns.boxplot(x="kind", y="pulse", data=exercise)
plt.title("Pulse Rate Distribution by Kind of Exercise")
plt.show()
```

**4. Use a pair plot to visualize relationships between numerical features.**

```python
sns.pairplot(exercise)
plt.show()
```

**5. Generate a swarm plot to show the distribution of pulse rates by kind of exercise and diet.**

```python
sns.swarmplot(x="kind", y="pulse", hue="diet", data=exercise)
plt.title("Pulse Rate Distribution by Kind of Exercise and Diet")
plt.show()
```

**6. Create a KDE plot to show the distribution of pulse rates for different kinds of exercises.**

```python
sns.kdeplot(data=exercise, x="pulse", hue="kind", fill=True)
plt.title("Pulse Rate Distribution by Kind of Exercise")
plt.show()
```
