```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")
```

**1. Display the first 5 rows of the dataset.**

```python
print(penguins.head())
```


 **2. Create a bar plot showing the count of different species.**

```python
sns.countplot(x="species", data=penguins)
plt.title("Count of Different Penguin Species")
plt.show()
```

**3. Generate a scatter plot showing flipper length vs body mass, colored by species.**

```python
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins)
plt.title("Flipper Length vs Body Mass by Species")
plt.show()
```

**4.  Create a box plot to show the distribution of body mass across different species.**

```python
sns.boxplot(x="species", y="body_mass_g", data=penguins)
plt.title("Body Mass Across Species")
plt.show()
```

**5. Plot a violin plot to visualize the distribution of flipper length across different species.**

```python
sns.violinplot(x="species", y="flipper_length_mm", data=penguins)
plt.title("Flipper Length Across Species")
plt.show()
```

**6. Use a pair plot to visualize relationships between numerical features.**

```python
sns.pairplot(penguins.dropna(), hue="species")
plt.show()
```

**7. Create a heatmap to show the correlation between numerical features.**

```python
corr = penguins.drop(columns=['species', 'island', 'sex']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Penguins Dataset")
plt.show()
```

**8. Generate a swarm plot to show the distribution of body mass by species and island.**

```python
sns.swarmplot(x="species", y="body_mass_g", hue="island", data=penguins)
plt.title("Body Mass by Species and Island")
plt.show()
```
