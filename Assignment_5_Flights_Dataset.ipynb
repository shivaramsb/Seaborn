```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset("flights")
```

**1. Display the first 5 rows of the dataset.**

```python
print(flights.head())
```


 **2. Create a line plot showing the number of passengers over time.**

```python
sns.lineplot(x="year", y="passengers", data=flights)
plt.title("Number of Passengers Over Time")
plt.show()
```

**3. Create a bar plot showing the average number of passengers per month.**

```python
sns.barplot(x="month", y="passengers", data=flights)
plt.title("Average Number of Passengers Per Month")
plt.show()
```

**4.  Plot a box plot to show the distribution of passengers each year.**

```python
sns.boxplot(x="year", y="passengers", data=flights)
plt.title("Passenger Distribution by Year")
plt.show()
```

**5. Use a swarm plot to show the distribution of passengers by year and month.**

```python
sns.swarmplot(x="year", y="passengers", hue="month", data=flights)
plt.title("Passenger Distribution by Year and Month")
plt.show()
```

**6. Create a factor plot to show the trend of passengers over months, separated by year.**

```python
sns.catplot(x="month", y="passengers", hue="year", kind="point", data=flights)
plt.title("Trend of Passengers Over Months, Separated by Year")
plt.show()
```

**7. Generate a point plot to visualize the number of passengers by month for a specific year.**

```python
sns.pointplot(x="month", y="passengers", data=flights[flights['year'] == 1960])
plt.title("Number of Passengers by Month for 1960")
plt.show()
```

**8. Create a catplot to show the distribution of passengers by year, colored by month.**

```python
sns.catplot(x="year", y="passengers", hue="month", kind="violin", data=flights)
plt.title("Passenger Distribution by Year, Colored by Month")
plt.show()
```
