```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
mpg = sns.load_dataset("mpg")
```

**1. Display the first 5 rows of the dataset.**

```python
print(mpg.head())
```


 **2. Create a bar plot showing the average miles per gallon (mpg) by origin.**

```python
sns.barplot(x="origin", y="mpg", data=mpg)
plt.title("Average MPG by Origin")
plt.show()
```

**3. Generate a scatter plot showing horsepower vs mpg, colored by origin.**

```python
sns.scatterplot(x="horsepower", y="mpg", hue="origin", data=mpg)
plt.title("Horsepower vs MPG by Origin")
plt.show()
```

**4.  Create a box plot to show the distribution of mpg across different cylinders.**

```python
sns.boxplot(x="cylinders", y="mpg", data=mpg)
plt.title("MPG Distribution by Cylinders")
plt.show()
```

**5. Plot a violin plot to visualize the distribution of weight across different origins.**

```python
sns.violinplot(x="origin", y="weight", data=mpg)
plt.title("Weight Distribution by Origin")
plt.show()
```

**6. Use a pair plot to visualize relationships between numerical features.**

```python
sns.pairplot(mpg.dropna())
plt.show()
```

**7. Generate a swarm plot to show the distribution of mpg by origin and cylinders.**

```python
sns.swarmplot(x="origin", y="mpg", hue="cylinders", data=mpg)
plt.title("MPG Distribution by Origin and Cylinders")
plt.show()
```

**8. Create a KDE plot to show the distribution of mpg for different origins.**

```python
sns.kdeplot(data=mpg, x="mpg", hue="origin", fill=True)
plt.title("MPG Distribution by Origin")
plt.show()
```
