```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
diamonds = sns.load_dataset("diamonds")
```

**1. Display the first 5 rows of the dataset.**

```python
print(diamonds.head())
```


 **2. Create a bar plot showing the count of diamonds by cut quality.**

```python
sns.countplot(x="cut", data=diamonds)
plt.title("Count of Diamonds by Cut Quality")
plt.show()
```

**3. Generate a scatter plot showing carat vs price, colored by clarity.**

```python
sns.scatterplot(x="carat", y="price", hue="clarity", data=diamonds)
plt.title("Carat vs Price, Colored by Clarity")
plt.show()
```

**4.  Create a box plot to show the distribution of prices across different cut qualities.**

```python
sns.boxplot(x="cut", y="price", data=diamonds)
plt.title("Price Distribution Across Cut Qualities")
plt.show()
```

**5. Plot a violin plot to visualize the distribution of carat across different colors.**

```python
sns.violinplot(x="color", y="carat", data=diamonds)
plt.title("Carat Distribution Across Colors")
plt.show()
```

**6. Use a pair plot to visualize relationships between numerical features.**

```python
sns.pairplot(diamonds.sample(1000))  # Sample for better performance
plt.show()
```
