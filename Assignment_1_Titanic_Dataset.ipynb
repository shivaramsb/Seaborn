```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")
```

**1. Display the first 5 rows of the dataset**

```python
print(titanic.head())
```


 **2. Create a bar plot showing the count of survivors based on passenger class**

```python
sns.countplot(x="class", hue="survived", data=titanic)
plt.title("Count of Survivors Based on Passenger Class")
plt.show()
```

**3. Plot a distribution of the ages of passengers**

```python
sns.histplot(titanic["age"].dropna(), kde=True)
plt.title("Distribution of Passenger Ages")
plt.show()
```

**4.  Create a box plot to show the distribution of fare prices across different passenger classes.**

```python
sns.boxplot(x="class", y="fare", data=titanic)
plt.title("Fare Prices Across Passenger Classes")
plt.show()
```

**5. Generate a heatmap of the correlation matrix of the dataset.**

```python
import pandas as pd

# Convert 'sex' and 'embarked' to numeric using one-hot encoding
titanic_numeric = pd.get_dummies(titanic, columns=['sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'], drop_first=True)

# Now compute the correlation matrix on the numeric DataFrame
corr = titanic_numeric.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Titanic Dataset")
plt.show()
```

**6. Create a pair plot to visualize the relationships between numerical features**

```python
sns.pairplot(titanic.select_dtypes(include=["float", "int"]).dropna())
plt.show()
```

**7. Use a violin plot to show the distribution of age across different classes, separated by gender.**

```python
sns.violinplot(x="class", y="age", hue="sex", data=titanic, split=True)
plt.title("Age Distribution Across Classes and Gender")
plt.show()
```

**8. Create a factor plot to show survival rate based on passenger class and gender.**

```python
sns.catplot(x="class", hue="sex", col="survived", data=titanic, kind="count")
plt.show()
```

**9.Generate a swarm plot showing fare prices by passenger class and survival status.**

```python
sns.swarmplot(x="class", y="fare", hue="survived", data=titanic)
plt.title("Fare Prices by Class and Survival Status")
plt.show()
```
