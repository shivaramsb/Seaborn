```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic = sns.load_dataset("titanic")
```

**1. Display the first 5 rows of the dataset.**

```python
print(titanic.head())
```


 **2. Create a joint plot showing the relationship between age and fare with a regression line.**

```python
sns.jointplot(x="age", y="fare", kind="reg", data=titanic)
plt.title("Age vs Fare with Regression Line")
plt.show()
```

**3. Generate a pair plot to visualize relationships between numerical features.**

```python
sns.pairplot(titanic.dropna())
plt.show()
```

**4.  Create a box plot to show the distribution of age across different classes, separated by survival status.**

```python
sns.boxplot(x="class", y="age", hue="survived", data=titanic)
plt.title("Age Distribution by Class and Survival Status")
plt.show()
```

**5. Plot a violin plot to visualize the distribution of fare across different classes, separated by gender.**

```python
sns.violinplot(x="class", y="fare", hue="sex", data=titanic, split=True)
plt.title("Fare Distribution by Class and Gender")
plt.show()
```

**6. Create a lmplot to show the relationship between age and fare, separated by class.**

```python
sns.lmplot(x="age", y="fare", hue="class", data=titanic)
plt.title("Age vs Fare by Class")
plt.show()
```

**7. Generate a bar plot showing survival rate by class and gender.**

```python
sns.barplot(x="class", y="survived", hue="sex", data=titanic)
plt.title("Survival Rate by Class and Gender")
plt.show()
```

**8. Create a strip plot to show the distribution of ages by class, colored by survival status.**

```python
sns.stripplot(x="class", y="age", hue="survived", data=titanic)
plt.title("Age Distribution by Class and Survival Status")
plt.show()
```
