```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")
```

**1. Display the first 5 rows of the dataset.**

```python
print(tips.head())
```


 **2. Create a joint plot showing the relationship between total bill and tip with a regression line.**

```python
sns.jointplot(x="total_bill", y="tip", kind="reg", data=tips)
plt.title("Total Bill vs Tip with Regression Line")
plt.show()
```

**3. Generate a pair plot to visualize the relationships between all numerical features.**

```python
sns.pairplot(tips)
plt.show()
```

**4.  Create a box plot to show the distribution of total bills across different time periods (Lunch/Dinner).**

```python
sns.boxplot(x="time", y="total_bill", data=tips)
plt.title("Total Bill Distribution by Time Period")
plt.show()
```

**5. Plot a violin plot to show the distribution of tips across different days, separated by gender.**

```python
sns.violinplot(x="day", y="tip", hue="sex", data=tips, split=True)
plt.title("Tip Distribution by Day and Gender")
plt.show()
```

**6. Create a lmplot to show the relationship between total bill and tip, separated by day.**

```python
sns.lmplot(x="total_bill", y="tip", hue="day", data=tips)
plt.title("Total Bill vs Tip by Day")
plt.show()
```

**7. Generate a bar plot showing average tip amount by day and gender.**

```python
sns.barplot(x="day", y="tip", hue="sex", data=tips)
plt.title("Average Tip Amount by Day and Gender")
plt.show()
```

**8. Create a strip plot to show the distribution of total bills by day, colored by time of day.**

```python
sns.stripplot(x="day", y="total_bill", hue="time", data=tips)
plt.title("Total Bill Distribution by Day, Colored by Time of Day")
plt.show()
```
