```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")
```

**1. Display the first 5 rows of the dataset.**

```python
print(tips.head())
```


 **2. Create a bar plot showing the average tip amount by day of the week.**

```python
sns.barplot(x="day", y="tip", data=tips)
plt.title("Average Tip Amount by Day of the Week")
plt.show()
```

**3. Generate a scatter plot of total bill vs tip, colored by time of day (lunch/dinner).**

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
plt.title("Total Bill vs Tip by Time of Day")
plt.show()
```

**4.  Create a box plot showing the distribution of total bills across different days.**

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Distribution Across Days")
plt.show()
```

**5. Plot a violin plot to show the distribution of tips by time of day.**

```python
sns.violinplot(x="time", y="tip", data=tips)
plt.title("Tip Distribution by Time of Day")
plt.show()
```

**6. Create a line plot showing the trend of average total bill amount over days of the week.**

```python
sns.lineplot(x="day", y="total_bill", data=tips, estimator='mean')
plt.title("Average Total Bill Amount by Day of the Week")
plt.show()
```

**7. Create a factor plot to show total bill amount by day and gender.**

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="bar", data=tips)
plt.title("Total Bill Amount by Day and Gender")
plt.show()
```

**8. Generate a swarm plot to show the distribution of tips by day, colored by gender.**

```python
sns.swarmplot(x="day", y="tip", hue="sex", data=tips)
plt.title("Tip Distribution by Day and Gender")
plt.show()
```
